```python
import pickle
from google.cloud import storage

bucket_name = "bucket_324041060"
blob_path = "title_id.pkl"   
client = storage.Client()
bucket = client.bucket(bucket_name)
blob = bucket.blob(blob_path)

data = blob.download_as_bytes()
obj = pickle.loads(data)

print("type:", type(obj))
```

```python
print("length:", len(obj))
print("first item:", obj[0])
print("type of first item:", type(obj[0]))
```

```python
import pickle
from google.cloud import storage

# obj כבר טעון ויש בו: list of Row(id=..., title=...)

clean_dict = {int(r.id): r.title for r in obj}

# שמירה כ-pickle מקומי
with open("title_id.pkl", "wb") as f:
    pickle.dump(clean_dict, f, protocol=pickle.HIGHEST_PROTOCOL)

print("Saved clean pickle with", len(clean_dict), "entries")
```

```python
client = storage.Client()
bucket = client.bucket("bucket_324041060")
blob = bucket.blob("title_id.pkl")  
blob.upload_from_filename("title_id.pkl")

print("Uploaded and overwritten title_id.pkl in bucket")
```
